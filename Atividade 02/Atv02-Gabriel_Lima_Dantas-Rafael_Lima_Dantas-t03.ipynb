```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
import scipy
from skimage.color import rgb2gray
from scipy.signal import convolve2d
from skimage.transform import rescale
```

```python
############## Código rescale.ipynb ##############

scale = 1/4
image = imread ('placa01.png',as_gray=True) # Aplicaçãp na imagem 1
image_restored= rescale(image, scale, anti_aliasing=True)
image_restored = rescale(image_restored, 1/scale, anti_aliasing=True)

_,ax = plt.subplots (1,3)
ax[0].imshow(image,cmap='gray')
ax[1].imshow(image_restored,cmap='gray')
ax[2].imshow(image_restored,cmap='gray')

# Adição para ver o gráfico:
plt.show()  # Exibe o gráfico
```

```python
############## Aplicação do filtro highboost ##############

# Filtro passa-baixas (filtro da média):
x = np.array([[1, 1, 1],
              [1, 8, 1], # Cria o filtro da média para uma janela de 3x3
              [1, 1, 1]]) / 9

filt_md = convolve2d (image_restored, x,  mode='same') # Faz a convolução da imagem com o filtro da média

# Criação da máscara:
# A imagem é a soma das altas e baixas frequências, como explicado em aula. Logo, para se obter as altas frequências, basta
# subtrair as baixas frequências da imagem

altas_frequencias = image_restored - filt_md

# Criação do fator peso da máscara do filtro Highboost

peso = 15 # Para essa imagem, foi encontrado a base de testes

# Multiplicação das altas frequências pelo peso da imagem

boost = peso * altas_frequencias # boost é a intensificação das altas frequências

# Soma da imagem com as altas frequências

image_highboost = image_restored + boost

image_highboost_= 255 - image_highboost # Aplica a operação do negativo na imagem
```

```python
# Exibição dos resultados
_, ax = plt.subplots(1, 3, figsize=(12, 4))
ax[0].imshow(image_restored, cmap='gray')
ax[0].set_title('Restaurada')
ax[1].imshow(filt_md, cmap='gray')
ax[1].set_title('Passa-Baixa (Filtro da Média)')
ax[2].imshow(image_highboost_, cmap='gray')
ax[2].set_title('Highboost (Filtro da Média)')
plt.show()
```
